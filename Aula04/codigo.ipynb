```python
import pandas as pd # importa a biblioteca pandas
```

```python
import matplotlib.pyplot as plt # importa a biblioteca matplolib
```

```python
dados = pd.read_csv('imigrantes_canada.csv') #importa base de dados
```

```python
dados.head() #exibe as 4 primeiras linhas
```

```python
dados.tail() #exibe as quatro ultimas linhas
```

```python
dados.info() #exibe as informações do dataframe
```

```python
dados.set_index(['País'],inplace=True)# seta a coluna como Pais como index
anos = list(map(str,range(1980,2013)))#transforma os anos m string utilizando função map
```

```python
anos
```

```python
brasil = dados.loc['Brasil',anos]#filtro para pegar os dados somente do Brasil
```

```python
brasil
```

```python
# criando dicionario com os dados brasil
Brasil_dic={
    'ano':brasil.index.tolist(), # paga os anos
    'imigrantes':brasil.values.tolist() #pega a quantidade de valores
}
```

```python
# criando dataframe para o brasil
df_brasil = pd.DataFrame(Brasil_dic)
```

```python
df_brasil.head() #exibe os imigrantes do dataframe brasil   
```

```python
#platando o numero de imigrantes do brasil para o canada
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])#plota o grafico
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes']) #plota o grafico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize = 14 ) #ajusta tamanho da fonte
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14 ) # ajusta valores para 
plt.xlabel(['Anos'],fontsize = 14 ) #ajusta eixo x
plt.ylabel(['Imigrantes do Brasil para Canadá'],fontsize = 18 ) #ajusta eixo y
plt.show()
```

```python
argentina = dados.loc['Argentina',anos]#filtro para pegar os dados somente do argentina
```

```python
Argentina_dic={
    'ano':argentina.index.tolist(), # paga os anos
    'imigrantes':argentina.values.tolist() #pega a quantidade de valores
}
```

```python
df_argentina = pd.DataFrame(Argentina_dic)
```

```python
plt.plot(df_argentina['ano'],df_argentina['imigrantes']) #plota o grafico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize = 14 ) #ajusta tamanho da fonte
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14 ) # ajusta valores para 
plt.xlabel(['Anos'],fontsize = 14 ) #ajusta eixo x
plt.ylabel(['Imigrantes Argentina p/ Canadá'],fontsize = 18 ) #ajusta eixo y
plt.show()
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'],color = 'red',label = 'Brasil')
plt.plot(df_argentina['ano'],df_argentina['imigrantes'],color = 'blue',label = 'Argentina')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize = 14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize = 14)
plt.title('Imigrantes Argentina p/ Canadá', fontsize=14)
plt.ylabel("N° de imigrantes",fontsize= 14)
plt.xlabel("Anos",fontsize = 14)
plt.legend()
plt.savefig("dados imigração.jpg")
plt.show()
```

```python
plt.bar(df_brasil['ano'],df_brasil['imigrantes']) #plt.bar
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize = 14)
plt.title('Imigrantes Argentina p/ Canadá', fontsize=14)
plt.show()


```

```python
df_comparacao=dados.loc[['Brasil','Argentina'],anos] #filtro que pega duas colunas do dataset
```

```python
df_comparacao # dataset de comparação
```

```python
df_comparacao = df_comparacao.T
df_comparacao
```

```python
df_comparacao = df_comparacao.T #aplica a matriz transposta no dataframe
df_comparacao['T']= df_comparacao['Brasil'] + df_comparacao['Argentina'] #soma o total de imigrantes dos dois dataframe
```

```python
plt.figure(figsize=(10,10))
plt.plot(df_comparacao['Brasil'],label='Brasil',color='green')
plt.plot(df_comparacao['Argentina'],label='Argentina',color='blue')
plt.plot(df_comparacao['T'],label='Total',color='red')
plt.legend()
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.title('Imigrantes Argentina p/ Canadá', fontsize=20)
plt.show()
```

```python
fig, ax  = plt.subplots(figsize=(8,4))  # criando uma figura
# ax - axes 
ax.plot(df_brasil['ano'],df_brasil['imigrantes'],lw=3,color = 'green',label='Brasil') # comando para plotar o grafico
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # altera o intervalo do eixo x
ax.xaxis.set_tick_params(labelsize=12) #altera o tamanho da fonte do grafico
ax.yaxis.set_tick_params(labelsize=12) #altera o tamanho da fonte do eixo y
ax.grid(linestyle='--')
ax.set_title('Imigração do Brasil para o Canadá  \n 1980 a 2013',loc='left')
ax.set_xlabel('Ano',fontsize=14)
ax.set_ylabel('Nº de imigrantes',fontsize=14)
ax.legend()
plt.show()
```

```python
# Criando a figura
fig, axs = plt.subplots(1, 2, figsize=(15,5)) # cria uma figura e armazena na variável axs
# axs - ax - é a abreviação de axes
# s é de subplot 

# Subplot 1: Gráfico de linhas
axs[0].plot(df_brasil['ano'], df_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')

axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
# Subplot 2: Boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Box-plot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()
```
