```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # Importa da biblioteca sklearn o modelo de Regressão Linear
```

```python
# 2) Criar um script para prever a temperatura de um forno a partir da quantidade de calor fornecida.

# Dados de TREINAMENTO:
Q = np.array([110,220,330,440,550,660,770]).reshape(-1, 1) # Quantidade de calor recebida
T = np.array([33,35,44,55,65,110,155]) # Temperatura resultante

#Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # Armazena a regressão linear na variável modelo
modelo.fit(Q,T) # Treina o modelo de regressão linear
```

```python
# Dados de teste
t_teste = np.array([100,200,300,400,500,600,700]).reshape(-1,1) 
#t_teste = np.array([30,35,40,45,60,100,150]).reshape(-1,1)

# Realizar a previsão com base nos dados fornecidos
temperatura_prevista = modelo.predict(t_teste)
```

```python
for i in range(len(t_teste)):
 print(f"Quantidade de calor: {t_teste[i][0]} C° Temperatura: {temperatura_prevista[i]} C°")
plt.scatter(Q, T, color='blue', label='Dados de Treinamento')
plt.plot(t_teste,temperatura_prevista, color = 'red',label='Dados da previsão')
plt.xlabel('Quantidade de calor fornecida (C°)')
plt.ylabel('Temperatura resultante (C°)')
plt.title('Relação entre Quantidade de calor fornecida e temperatura resultante')
plt.legend()
plt.grid(True)
plt.show()
```
