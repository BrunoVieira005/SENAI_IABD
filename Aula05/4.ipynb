```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # Importa da biblioteca sklearn o modelo de Regressão Linear
```

```python
# 4) Prever os gastos dos clientes com base no salário.

# Dados de TREINAMENTO:
Renda = np.array([1220,5005,8005,10005,30005]).reshape(-1,1) 
Gastos = np.array([755,1505,4005,6005,18005])

#Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # Armazena a regressão linear na variável modelo
modelo.fit(Renda, Gastos) # Treina o modelo de regressão linear
```

```python
# Dados para teste

teste = np.array([1200,5000,8000,10000,30000]).reshape(-1,1)

# Fazer previsões com base nos dados fornecidos
gastos_previstos = modelo.predict(teste)
```

```python
for i in range(len(teste)):
 print(f"Renda: R${teste[i][0]}  Gastos previsto: R${gastos_previstos[i]}")
plt.scatter(Renda, Gastos, color='blue', label='Dados de Treinamento')
plt.plot(teste, gastos_previstos, color = 'red',label='Dados da previsão')
plt.xlabel('Renda (R$)')
plt.ylabel('Previsão de gastos (R$)')
plt.title('Relação da Previsão entre Renda e Gastos')
plt.legend()
plt.grid(True)
plt.show()
```
