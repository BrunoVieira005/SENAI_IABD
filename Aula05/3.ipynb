```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # Importa da biblioteca sklearn o modelo de Regressão Linear
```

```python
# 3) Prever o valor de um imovel com base na área

# Dados de TREINAMENTO:
A = np.array([55,105,125,305]).reshape(-1,1) 
Valor = np.array([180.000,300.000,375.000,600.000])

#Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # Armazena a regressão linear na variável modelo
modelo.fit(A,Valor) # Treina o modelo de regressão linear
```

```python
# Dados para teste

A_teste = np.array([50,100,120,300]).reshape(-1,1)

# Realizar previsão com base nos dados fornecidos

valor_previsto = modelo.predict(A_teste)
```

```python
for i in range(len(A_teste)):
 print(f"Área: {A_teste[i][0]} m² Valor previsto: R${valor_previsto[i]} ")
plt.scatter(A, Valor, color='blue', label='Dados de Treinamento')
plt.plot(A_teste, valor_previsto, color = 'red',label='Dados da previsão')
plt.xlabel('Área (m²)')
plt.ylabel('Valor do imóvel (R$°)')
plt.title('Relação entre Área Construída e Valor do Imóvel')
plt.legend()
plt.grid(True)
plt.show()
```
