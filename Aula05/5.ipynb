```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # Importa da biblioteca sklearn o modelo de Regressão Linear
```

```python
# Criar um script para prever a demanda de energia com base no consumo dos meses anteriores pelos clientes 

consumo = np.array([55,141,33,77,125]).reshape(-1,1)
demanda= np.array([105,205,85,105,165]) 

#Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # Armazena a regressão linear na variável modelo
modelo.fit(consumo,demanda) # Treina o modelo de regressão linear
```

```python
# Dados para teste

teste = np.array([50,140,30,75,123]).reshape(-1,1)

# Fazer previsões com base nos dados fornecidos
demanda_prevista = modelo.predict(teste)
```

```python
for i in range(len(teste)):
 print(f"Consumo: {teste[i][0]}KW  Demanda prevista: R${demanda_prevista[i]}")
plt.scatter(consumo, demanda, color='blue', label='Dados de Treinamento')
plt.plot(teste, demanda_prevista, color = 'red',label='Dados da previsão')
plt.xlabel('Demanda')
plt.ylabel('Consumo (KW)')
plt.title('Relação da Previsão da Demanda de Energia por Consumo')
plt.legend()
plt.grid(True)
plt.show()
```
