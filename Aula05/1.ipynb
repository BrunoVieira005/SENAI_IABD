```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # Importa da biblioteca sklearn o modelo de Regressão Linear
```

```python
# 1) Criar um script para prever o nível de um tanque com vazão de 10,20,30,5,35,40 m3/ min

# Dados de TREINAMENTO:
vazao = np.array([12, 22, 33, 7, 37, 44]).reshape(-1, 1) # Vazão de água em m³/min
nivel = np.array([55, 65, 75, 40, 85, 88]) # Nível do tanque em metros

#Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # Armazena a regressão linear na variável modelo
modelo.fit(vazao, nivel) # Treina o modelo de regressão linear
```

```python
# Dados para previsão
vazao_teste = np.array([10, 20, 30, 5, 35, 40]).reshape(-1, 1) # vazão de água em m3/min
#vazao_teste = np.array([50, 60, 70, 45, 80, 85]).reshape(-1, 1) # vazão de água em m3/min

# Fazer previsões com base na vazão de água
nivel_previsto = modelo.predict(vazao_teste)
```

```python
# Impressão do gráfico

for i in range(len(vazao_teste)):
 print(f"Vazão: {vazao_teste[i][0]} m3/min - Nível previsto: {nivel_previsto[i]} metros")
plt.scatter(vazao, nivel, color='blue', label='Dados de Treinamento')
plt.plot(vazao_teste,nivel_previsto, color = 'red',label='Dados da previsão')
plt.xlabel('Vazão de Água (m3/min)')
plt.ylabel('Nível do Tanque (metros)')
plt.title('Relação entre Vazão de Água e Nível do Tanque')
plt.legend()
plt.grid(True)
plt.show()
```
