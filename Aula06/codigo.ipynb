```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split # Biblioteca para dividir os dados em Treino e Teste
from sklearn.tree import DecisionTreeClassifier # Importa o algoritmo da árvore de decisão
from sklearn.metrics import accuracy_score
from sklearn import tree # Importa o elemento tree para plotagem da árvore
```

```python
# Criando o conjunto de dados
notas = np.array([8,6,5,9,4,3,7,2,1,5,6,9,8])
classes = np.array(['bom','ruim','ruim','bom','pessimo','pessimo',
                    'bom','pessimo', 'pessimo','ruim','ruim','bom','bom'])
```

Dividindo a base de dados em treino e teste

```python
# Dividir dados em treino e teste
notas_treino, notas_teste, classes_treino, classes_teste = train_test_split(notas.reshape(-1,1),
                                                                            classes.reshape(-1,1),
                                                                            test_size=0.2,
                                                                            random_state=42) 
```

```python
modelo = DecisionTreeClassifier() # Cria o modelo
modelo.fit(notas_treino, classes_treino) # Treinamento do modelo
```

Teste do Modelo

```python
previsao = modelo.predict(notas_teste) # Dentro do modelo predict, são colocas as notas de teste
previsao
```

```python
# Teste de nota individual
nota = np.array([4,7]).reshape(-1,1)
resultado = modelo.predict(nota)
resultado
```

```python
fig = plt.figure(figsize=(10,8)) # Cria uma figura de tamanho 10x8
# Tree Plot plota a árvore de decisão
# Modelo é o modelo criado do algoritmo de aprendizado supervisionado
tree.plot_tree(modelo,feature_names=notas.tolist(), class_names=classes.tolist(), filled=True)
plt.title('Árvore de decisão para classificar uma bebida \n de acordo com a nota', fontsize=18)
plt.show()
```

```python
sns.boxplot(x= classes, y = notas)
plt.show()
```

```python
# gráfico de barras
plt.figure(figsize=(10,8))
sns.barplot(y=notas, x = classes)
plt.title('Notas de bebidas',fontsize=20)
plt.xlabel('Classificação',fontsize=16)
plt.ylabel('Notas',fontsize=16)
```
