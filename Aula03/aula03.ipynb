```python
import pandas as pd # Import na biblioteca Pandas
import numpy as np
import matplotlib.pyplot as plt 
```

```python
Dados_x = np.linspace(0,10,10) # Gera dados em x de 10 em 10 valores
Dados_y = Dados_x * Dados_x # Multiplica os dados de x e armazena em y
```

```python
plt.figure(figsize = (10,8), facecolor = 'gray') # Cria uma figura
Ax = plt.axes() # Cria os eixos para o gr√°fico
Ax.set_facecolor('white')
plt.plot (Dados_x, Dados_y, # Plota o gr√°fico em x e y
         color = 'red', marker = 'o',
         linestyle = 'dashed',
         markeredgecolor = 'black'
)
plt.xlabel("Eixo x", fontsize = 16) 
plt.ylabel("Eixo y", fontsize = 16)
plt.title("Gr√°fico simples com matplotlib", fontsize=20)

```

```python
plt.rcParams.update({'font.size': 15})  # Fun√ß√£o que permite ajustar a figura
plt.figure(figsize=(10,10))
linhas = 2 # Cria duas linhas
colunas = 2 # Cria duas colunas
plt.subplot(linhas, colunas, 1) # Subplot para plotar gr√°ficos um ao outro
plt.plot(Dados_x, 'r--', label='Dados_x') # Label r√≥tulo para os dados
plt.title("Gr√°fico 1")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend(loc ="best") # Cria a legenda na melhor posi√ß√£o dispon√≠vel
plt.subplot(linhas,colunas,2)
plt.plot(Dados_x, Dados_y, 'b-.', label = 'Dados x \n e y') # Plota o gr√°fico alterando a cor da linha
plt.title("Gr√°fico 2")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.subplot(linhas, colunas,3)
plt.plot(Dados_x * Dados_x, 'g-', label = "Dados x^2")
plt.title("Gr√°fico 3")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.subplot(linhas, colunas,4)
plt.plot(Dados_y * Dados_y, 'c--')
plt.title("Gr√°fico 4")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.show()
```

```python
# Gr√°fico de barras
Figura = plt.figure()
Grafico = Figura.add_axes([0,0,0.7,0.7]) # Cria os eixos para o gr√°fico
Rotulos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
# Gerando gr√°fico
Grafico.bar(Rotulos, Dados_x)
```

```python
Grafico.bar(Rotulos,Dados_x, color = 'r')
Grafico.bar(Rotulos, Dados_y*0.1, color='b')
Figura
```

```python
Grafico.bar(Dados_x, Dados_y/2, color = ['red', 'blue']) # Exibe os dados um do lado do outro
Figura
```
